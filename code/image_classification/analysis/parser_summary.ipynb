```python

# Read csv file
df_1 = pd.read_csv('summary_1.csv')
df_10 = pd.read_csv('summary_10.csv')
df_100 = pd.read_csv('summary_100.csv')

# Remove column "Hierarchy Level" from all the dataframes

df_1 = df_1.drop(['Hierarchy Level'], axis=1)
df_10 = df_10.drop(['Hierarchy Level'], axis=1)
df_100 = df_100.drop(['Hierarchy Level'], axis=1)


# join the two dataframes on the basis of column "Metric Name"

df_1_10 = pd.merge(df_1, df_10, on='Metric Name', how='inner')
df_1_10_100 = pd.merge(df_1_10, df_100, on='Metric Name', how='inner')

# Rename the columns to 1 , 10 and 100

df_1_10_100.columns = ['Metric Name', '1', '10', '100']

```

```python
# Plot df_1_10_100 dataframe

import matplotlib.pyplot as plt
```

```python
# create one plot for each row and save it in the same directory

for i in range(0, len(df_1_10_100)):
    df_1_10_100.iloc[i:i+1, 1:].plot(kind='bar', title=df_1_10_100.iloc[i:i+1, 0:1].values[0][0])
    plt.savefig(df_1_10_100.iloc[i:i+1, 0:1].values[0][0] + '.png')
    plt.show()
```
